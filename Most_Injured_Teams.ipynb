```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load your dataset into a pandas DataFrame
df = pd.read_csv('injury_data_updated copy.csv')
print(df.head())
```

```python
injury_counts_by_team = df.groupby('Team Name')['Injury Type'].count()
print(injury_counts_by_team)
```

```python
# Identify the team with the highest number of injuries
most_injuries_team = injury_counts_by_team.idxmax()
most_injuries_count = injury_counts_by_team.max()
print(f"Position with the highest injuries: {most_injuries_team} ({most_injuries_count} injuries)")
```

```python
# Group the data by the 'Team' column and count the number of injuries
injury_counts_by_team = df.groupby('Team Name')['Injury Type'].count().reset_index()
```

```python
# Rename columns
injury_counts_by_team.columns = ['Team Name', 'Injury Type']
```

```python
# Create a Bar Chart   
plt.figure(figsize=(35, 10))
plt.bar(injury_counts_by_team['Team Name'], injury_counts_by_team['Injury Type'], color='pink')

# Add labels and title
plt.xlabel('Teams', fontsize=30)
plt.ylabel('Number of Injuries', fontsize=30)
plt.title('Injuries per Team', fontsize=25)
plt.xticks(rotation=45, ha='right',fontsize=25)

# Sort the DataFrame by Injury Type in descending order
injury_counts_by_team = injury_counts_by_team.sort_values(by='Injury Type', ascending=False)

# Display the bar chart
plt.tight_layout()
plt.show()
```

```python
# Create alt Chart
plt.figure(figsize=(10, 6))
plt.barh(injury_counts_by_team['Team Name'], injury_counts_by_team['Injury Type'], color='skyblue')
plt.xlabel('Injuries')
plt.ylabel('Team')
plt.title('Number of Injuries per Team')
plt.gca().invert_yaxis()
plt.show()
```

```python
# Calculate total injuries per week
injuries_by_week = df.groupby('Week').size()

# Plotting a bar chart
plt.figure(figsize=(10, 6))
plt.bar(injuries_by_week.index.astype(int), injuries_by_week.values, color='skyblue')
plt.xlabel('Week')
plt.ylabel('Number of Injuries')
plt.title('Total Injuries per Week')
plt.xticks(injuries_by_week.index.astype(int))
plt.tight_layout()
plt.show()

# Week with the highest number of injuries
highest_injuries_week = injuries_by_week.idxmax()
print(f"Week with the highest number of injuries: {highest_injuries_week}")

# Calculate average injuries per week
avg_injuries_per_week = injuries_by_week.mean()
print(f"Average injuries per week: {avg_injuries_per_week}")
```
